### 1. Player Attributes Correlation:

```python
import seaborn as sns
import pylab as pl
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates
%matplotlib inline
```

```python
data = pd.read_csv('plot_patterns.csv')
data_corr = data.loc[:, ['Free kick accuracy', 'Heading accuracy', 'Interceptions', 'Jumping', 'Long passing', 'Long shots', 'Marking', 'Penalties', 'Positioning', 'Reactions', 
                 'Short passing', 'Shot power', 'Sliding tackle', 'Sprint speed', 'Stamina', 'Standing tackle', 'Strength', 'Vision', 'Volleys']]
corr = data_corr.corr()
sns.heatmap(corr)
```

From the above plot, we make few important observations:
    1. Free kick accuracy is strongly related to Long shots and Volleys.
    2. Interceptions have a strong correlation with Marking and Sliding & standing tackle.
    3. Long shots has a direct relation with Positioning and Penalties. So, if a player has good stats in Long shots, then it is highly likely that he may have a good sense of position too and is a good penalty kick taker.

### 2. Player Attributes Relating to position:
#### 2.1. Forward position:

```python
data = pd.read_csv('plot_patterns.csv')
data['Pos'] = 'FWD'
df_st = data['Preferred Positions'].str.contains('ST|LW|RW|LS|RS|CF')
p = (data[df_st]
         .loc[data['Overall']> 70, ['Acceleration', 'Aggression', 'Agility', 
                                    'Balance', 'Ball control', 'Composure', 
                                    'Crossing', 'Dribbling', 'Finishing', 'Pos']]
    )
parallel_coordinates(p, 'Pos')
```

One of the most astonishing insight here is highly variable "Acceleration", "Aggression", and "Finishing". For a forward with Overall > 70, there is a certain threshold for these attributes since they are vital for this position. And the "Ball Control" is less varying among these players which makes sense as they need high control over the ball in the tight spaces.

#### 2.2. Midfield Position:

```python
data = pd.read_csv('plot_patterns.csv')
data['Pos'] = 'MID'
df_st = data['Preferred Positions'].str.contains('CM|CAM|CDM|LM|RM')
p = (data[df_st]
         .loc[data['Overall']> 70, ['Acceleration', 'Balance', 'Ball control', 'Composure', 
                                    'Short passing', 'Interceptions', 'Long shots', 'Strength', 'Vision', 'Pos']]
    )
parallel_coordinates(p, 'Pos')
```

This analysis gives us a good idea about high variance in "Interception" attribute followed by "Long shots" and "Strength". This makes sense because we have considered Agressive and Defensive midfield players as well. 

#### 2.3. Defence Position:

```python
data = pd.read_csv('plot_patterns.csv')
data['Pos'] = 'DEF'
df_st = data['Preferred Positions'].str.contains('CB|LB|RB')
p = (data[df_st]
         .loc[data['Overall']> 70, ['Acceleration', 'Heading accuracy', 'Jumping', 
                                    'Short passing', 'Interceptions', 'Long passing', 
                                    'Long shots', 'Strength', 'Vision', 'Pos']]
    )
parallel_coordinates(p, 'Pos')
```

In the above plot we can understand that the "Strength", "Heading accuracy", "Interceptions" attributes have least variance. This makes sense because these are vital for a defender.
